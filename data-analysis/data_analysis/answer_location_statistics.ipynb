# Extract statistics about the location of the correct sentences in the original test

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_pickle("../data_frames/parsed_answer_data/df_train_cleaned.pkl")
```

```python
def plot_locations(df, title, bins):
    locations = df['answer_location'].values
    min = np.min(np.array(locations))
    max = np.max(np.array(locations))
    print('min: ', min)
    print('max: ', max)
    bins = list(range(min, max+1))
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, bins=bins, color='teal')
    plt.ylabel('Number of Q/A pairs')
    plt.xlabel('Location (sentence index)')
    plt.title(title)
    plt.show()
```

```python
# create histogram of in which sentence the answer appear
def plot_normalized_locations(df, title, bins):
    locations = df['loc_frac'].values
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, bins=bins, color='teal')
    plt.ylabel('Number of Q/A pairs')
    plt.xlabel('Location (relative)')
    plt.title(title)
    plt.show()
```

```python
# plot training data
plot_locations(df_train, 'Location of sentence with correct answer', 100)
plot_normalized_locations(df_train, 'Relative location of sentence with correct answer', 10)
```
