# Extract statistics about the location of the correct sentences in the original test

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_pickle("../data_frames/parsed_answer_data/df_train_cleaned.pkl")
```

```python
def plot_locations(df, title, bins):
    locations = df['answer_location'].values
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, bins=bins)
    plt.ylabel('Num sentences')
    plt.xlabel('Location (sentence index in paragraph)')
    plt.title(title)
    plt.show()
```

```python
# create histogram of in which sentence the answer appear
def plot_normalized_locations(df, title, bins):
    locations = df['loc_frac'].values
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, bins=bins)
    plt.ylabel('Num sentences')
    plt.xlabel('Location (as fraction)')
    plt.title(title)
    plt.show()
```

```python
# plot training data
plot_locations(df_train, 'Location of sentence with correct answer', 100)
plot_normalized_locations(df_train, 'Location of sentence with correct answer as fraction', 100)
```
