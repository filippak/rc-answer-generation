# Extract statistics about the location of the correct sentences in the original test

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_pickle("./df_train.pkl")
df_test = pd.read_pickle("./df_test.pkl")
```

```python
def plot_locations(df, title, bins):
    locations = df['answer_location'].values
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, density=True, bins=bins)
    plt.ylabel('Vol')
    plt.xlabel('Location (sentence index in paragraph)')
    plt.title(title)
    plt.show()
```

```python
# create histogram of in which sentence the answer appear
def plot_normalized_locations(df, title, bins):
    locations = df['loc_frac'].values
    # flat_list = [item for sublist in locations for item in sublist]
    plt.hist(locations, density=True, bins=bins)
    plt.ylabel('Vol')
    plt.xlabel('Location (as fraction)')
    plt.title(title)
    plt.show()
```

```python
# plot training data
plot_locations(df_train, 'Training data', 100)
plot_normalized_locations(df_train, 'Traning data, as fraction', 50)
```

```python
# plot test data
plot_locations(df_test, 'Test data', 60)
plot_normalized_locations(df_test, 'Test data, as fraction', 50)
```
