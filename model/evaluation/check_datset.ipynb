```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '../../data-analysis/create_dataset/data/CA/labeled_CA_data_train.pkl'
df = pd.read_pickle(path)
for idx, row in df.iterrows():
    labels = row['labels']
    tokens = row['tokens']
    for idx, label in enumerate(labels):
        if label == 1:
            l = tokens[idx]
            count = 1
            while idx+count < len(labels):
                next_label = labels[idx+count]
                if next_label != 2:
                    break
                l += ' '+tokens[idx+count]
                count += 1
            print('answer: ', l)

```
